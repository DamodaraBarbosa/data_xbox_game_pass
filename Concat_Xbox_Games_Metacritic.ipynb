# Concatenando os dataframes (Xbox One e Xbox Series) com notas no Metacritic


Importando as bibliotecas

```python
import pandas as pd
import numpy as np
```

## Definindo os Dataframes

Com os dados de jogos do Xbox 360

```python
df_xbox_360_metacritic = pd.read_csv('/content/drive/MyDrive/Dados Xbox/df_xbox_360_metacritic.csv')
```

```python
df_xbox_360_metacritic.head()
```

Com os dados de jogos do Xbox One

```python
df_xbox_one_metacritic = pd.read_csv('/content/drive/MyDrive/Dados Xbox/metacritic_xbox_one.csv')
```

```python
df_xbox_one_metacritic.head()
```

Com os dados de jogos do Xbox Series X

```python
df_xbox_series_metacritic = pd.read_csv('/content/drive/MyDrive/Dados Xbox/metacritic_xseries (1).csv')
```

```python
df_xbox_series_metacritic.head()
```

## Concatenando os Dataframes

```python
xbox_games_metacritic = pd.concat([df_xbox_360_metacritic, df_xbox_one_metacritic, df_xbox_series_metacritic], 
                            ignore_index= True)
```

```python
xbox_games_metacritic.head()
```

```python
xbox_games_metacritic.shape
```

# Tratando o dataframe

```python
# vericando os jogos que se repetem:

repeated_games = xbox_games_metacritic.duplicated(subset= 'Name')
```

```python
xbox_games_metacritic[repeated_games]
```

```python
# alguns jogos foram lançados em diferentes gerações dos consoles Xbox. Portanto, não é interessante manter dados
# que se repetem. Nesse projeto dar-se-á mais importância para os dados mais recentes, ou seja, do Xbox Series.

xbox_games_metacritic.drop_duplicates(subset= 'Name', keep= 'last', ignore_index= True,
                                  inplace= True)
```

```python
xbox_games_metacritic.shape
```

## Exportando o Dataframe

```python
xbox_games_metacritic.to_csv('xbox_games_metacritic.csv', index = False, encoding = 'utf-8-sig')
```
